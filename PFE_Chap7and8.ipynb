```python
def chop(t):
    try:
        del t[0]
        del t[-1]
        return None
    except:
        print ('Not a list')

g = [1, 2, 3, 4, 5]

h = '12345'

print (g)
chop (g)
print (g)

print (h)
chop (h)
print (h)
```

```python
def middle(t):
    try:
        g = t[1:-1]
        return g
    except:
        print ('Not a list')

a = [1, 2, 3, 4, 5]

b = '12345'

print (a)
c = middle(a)
print (c)

print (b)
d = middle (b)
print (d)
```

```python
fhand = open('mbox.txt')
count = 0
for line in fhand:
    if 'From ' not in line or len(line) == 0:
        continue
    else:
        words = line.split()
        if words[0] == 'From':
            #print ('Debug:', words)
            print(words [1])
            count = count + 1
        
print ('There were ' + str(count) + ' lines in the file with From as the first word')
```
